```python
za_god = za_god.rename(columns={'Container_Type':'Тип контейнера','cnt':'Количество'})
za_q = za_q.rename(columns={'Container_Type':'Тип контейнера','cnt':'Количество'})
za_vse = za_vse.rename(columns={'Container_Type':'Тип контейнера','cnt':'Количество'})
```

Посмотрим распределение различных типов контейнеров за последний (3) квартал, за 2023 год и за все время

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

labels0 = za_q['Тип контейнера']
labels = za_god['Тип контейнера']
labels1 = za_vse['Тип контейнера']

# Create subplots: use 'domain' type for Pie subplot
fig = make_subplots(rows=1, cols=3, specs=[[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}]])
fig.add_trace(go.Pie(labels=labels0, values=za_q['Количество'], name="Q3"),
              1, 1)
fig.add_trace(go.Pie(labels=labels, values=za_god['Количество'], name="2023 Год"),
              1, 2)
fig.add_trace(go.Pie(labels=labels1, values=za_vse['Количество'],  pull=[0, 0, 0, 0, 0, 0, 0.1], name="За все время"),
              1, 3)

# Use `hole` to create a donut-like pie chart
fig.update_traces(hole=.4, hoverinfo="label+value+percent+name")

fig.update_layout(
    title_text="Распределение типов контейнеров за разные периоды времени",
    # Add annotations in the center of the donut pies.
    annotations=[dict(text='Q3', x=0.135, y=0.5, font_size=17, showarrow=False),
                 dict(text='2023 Год', x=0.5, y=0.5, font_size=17, showarrow=False),
                 dict(text='За все время', x=0.895, y=0.5, font_size=17, showarrow=False)])
fig.show()
```

**Замечание**: Не заполненные данные есть только до 2023 -> Полнота данных за **Q3 и 2023 = 100%**
